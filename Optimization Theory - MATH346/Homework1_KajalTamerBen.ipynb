```python
using Gadfly
```

```python
f(x) = 8sinh(x/4) - (x^2) / 2
```

```python
plot(f,-40, 40)
```

```python
df(x) = 2cosh(x/4) - x
d2f(x) = 0.5sinh(x/4) -1
```

```python
function newton(f, df, d2f, x; maxIter=20, tol=1e-10)
his = zeros(maxIter)
i = 1
for i = 1:maxIter
    his[i] = x
    x = x - df(x)/d2f(x)
    if abs(df(x)/d2f(x)) <= tol
        break
    end
end
return x, his[1:i]
end

    
    
    
```

```python
for x = (0, 3, 40)
println(newton(f, df, d2f, x))
end
```

```python
for i = (0, 3, 40)
x, h = newton(f, df, d2f, i)
d2x = d2f(x)
if d2x == 0
    break   
elseif(d2x) < 0
    println("f(x) has a local maximum at x=",x)
else
    println("f(x) has a local minimum at x=",x)
end
end
```

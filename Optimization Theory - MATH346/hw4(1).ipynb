We use the gradient descent function from the class google drive folder:

```julia
include("/home/juser/math346/optiFuncs.jl")

T = [0.9, 1.5, 13.8, 19.8, 24.1, 28.2, 35.2, 60.3, 74.6, 81.3]
C = [455.2, 424.6, 122.1, 68.3, 43.2, 26.1, 13.1, 1.4, 0.4, 0.3]

m(x, t) = x[1] * exp(x[2] * t) + x[3]
dm(x, t) = [exp(x[2] * t)
            x[1] * t * exp(x[2] * t)
            1]
    
f(x) = 0.5 * sum([(m(x, T[i]) - C[i])^2 
                  for i=1:10])
df(x) = sum([(m(x, T[i]) - C[i]) * dm(x, T[i]) for i=1:10])

x0 = [100; -1; 0]
myLS(f,fc,df,x,pk) =  armijo(f,fc,df,x,pk,alpha=1e-1,b=0.7,maxIter=500)
gd(f, df, x0, maxIter=500, ls=myLS)
#f(gd(f, df, x0, maxIter=500, ls=myLS)[1])

```

Gauss Newton method:

```julia
d2m(x, t) = [0                          t*exp(x[2] * t)   0
             t^2 * x[1] * exp(x[2] * t) t * exp(x[2] * t) 0
             0                          0                 0]
d2f(x) = sum([m(x, T[i]) * d2m(x,T[i]) + dm(x, T[i]) * dm(x, T[i])' - C[i] * d2m(x, T[i])
    for i=1:10])

newton(f, df, d2f, x0)
#f(newton(f, df, d2f, x0, maxIter=500000)[1])
```

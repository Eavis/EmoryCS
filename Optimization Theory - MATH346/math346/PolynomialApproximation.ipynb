## Polynomial Best Approximation

We want to construct a polynomial of degree $n$ that best approximates $\cos(2 \pi x)$ on the interval $[0,1]$. To this end, we discretize $g(x)$ at points $x_0, x_1, \ldots, x_{20}$ in $[0,1$ to get
\begin{equation}
    g(x) = \left(\begin{array}{r} g(x_0)\\g(x_1) \\ \vdots \\ g(x_{20}) \end{array}\right).
\end{equation}
For a given $n=2,3,4,5$ we construct the extended Vandermonde matrix
\begin{equation}
    A(x) = \left(\begin{array}{rrrr} 
                            1 & x_0 &\ldots & x_0^n \\
                            1 & x_1 &\ldots & x_2^n \\
                            \vdots & \vdots &\ldots & \vdots \\
                            1 & x_{20} &\ldots & x_{20}^n
                     \end{array}\right).
\end{equation}
and solve
$$
    \min_{c\in\mathbb{R}^{n+1}} f(c) \quad \text{ where } \quad f(c) = \frac{1}{2} \| A c - g \|_2^2.
$$

Solutions to this least-squares problem are obtained from the normal equations
$$
    \nabla f(c) = A^T (A c - g) = 0 \Rightarrow c = (A^T A ) ^{-1} (A^T g).
$$
Since $f$ is convex, every stationary point is a global minimzer. Further, if $A$ has full rank, $f$ is strictly convex and there is a unique solution.

## Solve the least squares problem  for different $n$

```python
# get test data
g(x)   = cos(2*pi*x)
x      = linspace(0,1,21)
gd     = g(x)
p(x,c) = c[1] + c[2]*x + c[3]*x.^2 + c[4]*x.^3 + c[5]*x.^4 + c[6]*x.^5

# choose all N's to compute
N = [2,3,4,5,]

# Build Vandermonde matrix
A = [ones(length(x)) x[:] x[:].^2 x[:].^3  x[:].^4  x[:].^5]
C = zeros(6,length(N))
E = zeros(6,length(N))
for k=1:length(N)
    n = N[k]
    At = A[:,1:n+1]
    
    c = (At'*At)\(At'*gd)
    C[1:n+1,k] = c
    
    e, = eig(At'*At)
    E[1:n+1,k] = e
end
```

## Plot the Eigenvalues and Condition Number for each $n$

```python
subplot(1,2,1)
semilogy(N,E[1,:]',hold=true)
semilogy(N,maximum(E,1)',hold=true)
legend(("min eigenvalue", "max eigenvalue"),loc=3)
title("eigenvalues")
xlabel("n")
subplot(1,2,2)
semilogy(N,(maximum(E,1)./E[1,:])',hold=true)
title("condition number")
xlabel("n")
```

The figure above shows the eigenvalues of $A^T A$ (left) and its condition number (right) for each $n=2,3,4,5$. It can be seen that the smalles eigenvalue becomes very small already for $n=5$ and the condition number, computed as
$$
    \kappa(A) = \frac{\lambda_{max}(A)} {\lambda_{min}(A)}
$$
becomes very large. This implies
1. Since the smallest eigenvalue is always positive, $A^T A$ is positive definite and the solution to the least-squares problem is uniquely defined.
1. Since the smallest eigenvalue becomes very small, $A^T A$ is almost singular. Although $\lambda_{min}$ is always positive in theory, it might be so small that we experience numerically instabilities.
1. Recall, that a large condition number implies that small pertubations of the right hand side lead to large perturbations in the solution of 
$$
A x = b.
$$


## Plot the approximations and absolute errors

```python
xt = linspace(0,1,201)
plot(xt,g(xt),hold=true)
for k=1:length(N)
    pt = p(xt,C[:,k])
    plot(xt,pt)
    err = maximum(abs(pt-g(xt)))
    @printf "n=%3d\t max.abs.error=%1.6f\n" N[k] err
end
legend(("g(x)","p2(x)","p3(x)","p4(x)","p5(x)"))
```

The figure above shows the true function and the approximating polynomials $p_n$. Also the maximum absolute error is printed. There are two observations:
1. Clearly the approximation becomes better for larger $n$
1. The error of $p_2$ and $p_3$ as well as $p_4$ and $p_5$ are the same. This is due to the fact that
\begin{equation*}
					\cos(x) = \sum_{k=0}^\infty \frac{x^{2k}}{2k!}
\end{equation*}			

## Adding noise to the data

We now repeat the above experiment, but perturb the right hand side by some random noise. Here, we restrict the discussion to $n=4,5$, which should in theory behave exactly the same.

```python
gd     = g(x)
nd     = randn(length(x))
sigma  = logspace(2,-4,51)

# Build Vandermonde matrix
A4 = A[:,1:5]
A5 = A[:,1:6]
C4 = zeros(6,length(sigma))
C5 = zeros(6,length(sigma))
E4 = zeros(length(sigma))
E5 = zeros(length(sigma))

for k=1:length(sigma)
    c4 = (A4'*A4)\(A4'*(gd+sigma[k]*nd))
    C4[1:5,k] = c4
    E4[k] = maximum(abs(g(xt)-p(xt,C4[:,k])))
    
    c5 = (A5'*A5)\(A5'*(gd+sigma[k]*nd))
    C5[:,k] = c5
    E5[k] = maximum(abs(g(xt)-p(xt,C5[:,k])))
end
```

```python
subplot(1,2,1)
loglog(sigma,E4,hold=true)
loglog(sigma,E5)
title("maximum absolute error")
xlabel("noise level")
legend(("error for p4", "error for p5"),loc=2)
subplot(1,2,2)
sig = 24;
plot(xt,g(xt),hold=true)
plot(xt,p(xt,C5[:,sig]))
plot(x,gd+sigma[sig]*nd,"or")
title("sigma=$(sigma[sig])")
xlabel("x")
legend(("g(x)", "p5","data"),loc=2)
```

The above figure shows the maximum error depending on the noise level (left) and for a particular noise level (right). It can be seen that
1. For both $n$ the approximation becomes worse for larger noise levels. Also, both approximations are affected roughly in a similar fashion. 
1. Even for very strong noise, we can approximate $g$ quite well. So, the method is relatively robust.

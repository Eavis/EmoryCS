# Plotting functions in 2D

This notebook provides some basic plotting functionality for 2D functions in Julia. It should be a useful tool to analyze optimization problems in two variables.

```julia
using PyPlot
```

```julia
# some quadratic function
V = [1 1; -1 1]; S = [1 0; 0 -1]; A = V'*S*V;
f(x1,x2) = 0.5*[x1 x2]*A*[x1,x2]
x1 = linspace(-10,10,101);
x2 = linspace(-10,10,101);
fc = [f(x1[i],x2[j])[1] for i=1:101,j=1:101];
```

```julia
# the objective function of the Coffee production problem in the lecture
f(x1,x2) = (3+1)*x1 + (5+1)*x2 - (x1*(5-0.002*x1+ 0.001 *x2) + x2*(8-0.002*x2 + 0.001*x1))
x1 = linspace(0,1500,101);
x2 = linspace(0,2000,101);
fc = [f(x1[i],x2[j]) for i=1:101,j=1:101];
```

```julia
# some other function
V = [1 1; -1 1]; S = [1 0; 0 2]; A = V'*S*V;
f(x1,x2) = 0.5*[x1 x2]*A*[x1,x2]
x1 = linspace(-10,10,101);
x2 = linspace(-10,10,101);
fc = [f(x1[i],x2[j])[1] for i=1:101,j=1:101];
```

## Draw contour lines of f

```julia
contour(x1,x2,fc,30)
ylabel("x2")
xlabel("x1")
colorbar()
```

## Surface plot

```julia
surf(x1,x2,fc)
ylabel("x2")
xlabel("x1")
zlabel("f(x1,x2)")
```

## Plot as image

```julia
matshow(fc)
ylabel("x2")
xlabel("x1")
```

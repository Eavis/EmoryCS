## Our data

This example considers the following data fitting problem: Given measurements of the concentration of a new medicine in a patient's blood stream over time, we want to calibrate the coefficients of the model
$$
\phi(x; t) = x_1 + t x_2 + t^2 x_3^2 + x_4 \exp(-x_5 t)
$$
to predict the data.

Before we begin, let us take a look at the data, and predictions obtained from some paramters x0.

```python
td = [1.0; 1.8; 2.1; 2.25; 2.8; 3.5; 4; 5; 6]
cd = [10; 7.0; 6; 5; 2.6; 2.4; 2; 1.9;1.8]

phi(x,t) = x[1] + t.*x[2]+x[3].*t.^2 + x[4].*exp(-x[5].*t)
x0 = [1;.02;0.01;15;.5]

t = linspace(1,6,101)
plot(t,phi(x0,t),hold=true)
plot(td,cd,"or")
legend(("model prediction","data"))
```

It is obvious that the current model, does not predict the data very well (see next Section for a more solid argument). Thus, our goal is to improve the quality of the model through optimization. We seek to solve the nonlinear least squares problem
$$
 \min_x \frac{1}{2} \sum_{i=1}^N \| r_i(x)  \|^2 \text{ where } r_i(x) = \phi(x; t_i) - d_i
$$

## Show that starting guess is not optimal. 

Let us first compute the objective function and derivatives

```python
f(x) = 0.5*norm(phi(x,td)-cd)^2
dr(x) = [ones(length(td)) td td.^2  exp(-x[5]*td)  -x[4].*td.*exp(-x[5]*td)]'
r(x) = phi(x,td)-cd
df(x) = dr(x)*r(x);
```

Having computed the gradient of the objective function, we can check if our starting guess, x0, is optimal. Recall, that a sufficient first-order condition is that x0 is stationary, i.e.,
$$
 \| \nabla f(x_0) \| = 0.
$$
In our case we have

We can now check for optimality

```python
 norm(df(x0))
```

So, this proves our intuition from above namely that x0 is not optimal. 

## Try to improve our starting guess!

In order to improve our model prediction, perform two steps. First, we find a direction $\Delta x$ in which $f$ decreases. Then, we find a step size $t$ by solving
$$
\min_{t>0} f(x_0 + t \Delta x).
$$

Taking a look at the first order Taylor series as x0, we see that a condition on $\Delta x$ to be a descent direction is
$$
f(x_0 + t \Delta x) = f(x_0 ) + t \nabla f(x_0)^T \Delta x + \mathcal{O}(t^2) \Rightarrow \nabla f(x_0)^T \Delta x  < 0
$$ 
this is the so-called angle condition. 

One choice that always fulfills the angle conditions is
$$
\Delta x = - \nabla f(x_0)
$$
which leads to the so called Gradient Descent method.

```python
@printf "i=%4d, f=%1.2f |df|=%1.1e \n" 0 f(x0) norm(df(x0))
xgd = copy(x0)
stepMax = 0.002
for i=1:200
    # pick descent direction
    dx = -df(xgd)

    # find out how far to go
    tt = linspace(0,stepMax,201)
    ft = [f(xgd+tt[k]*dx) for k=1:length(tt)]
    id = indmin(ft)
    t0 = tt[id]
    
    # update model
    xgd = xgd+t0*dx
    if mod(i,10)==0
        @printf "i=%4d, f=%1.2f |df|=%1.1e, t=%1.1e \n" i f(xgd) norm(df(xgd)) t0
    end
end
plot(t,phi(xgd,t),hold=true)
plot(td,cd,"or")
legend(("Gradient Descent solution","data"));
```

We see that the Gradient Descent method improved the quality of our model prediction with respect to the nonlinear least squares objective function. Also, the norm of the gradient decreased significantly. However, it is quite tricky to the choice of ``stepMax``. Play around with it!

## Improve accuracy using Gauss Newton / Levenberg Marquardt method

There are two shortcomings associated with the Gradient Descent method. First, we have to perform many iterations before we reach a sufficiently good reconstruction. Second, it is very difficult to find a good value for ``stepMax``. 

One way to overcome these drawbacks is by solving a series of linear(!) least squares problems. These, as we have seen, are relatively easy to solve. 

In order to obtain a linear least squares problem, we apply Taylor's Theorem to the residual and obtain the problem

$$
\Delta x_1 = \arg\min_{\Delta x} 0.5 \| \nabla r(x_0)^\top \Delta x  + r(x_0) \|^2.
$$
The first order necessary condition then is
$$
\nabla r(x_0) \nabla r(x_0)^\top  \Delta x_1 = - \nabla r(x_0)^\top   r(x_0)
$$
which is well-defined if $\nabla r(x_0)$ has full rank. We then update the coefficients by
$$
    x_1 = x_0 + t \Delta x_1
$$
and, as we will see, the optimal line search parameter $t$ will be close to 1. 



When the rank of $\nabla r(x_0)$ is not full, the above iteration is not well-defined and the norm of $\Delta x$ might be very large. This is a well-known drawback of Gauss-Newton methods but easy to fix. The idea of the Levenberg Marquardt method is to limit the step to a so-called Trust Region. This yields the following problem
$$
\min_{\Delta x} 0.5 \| \nabla r(x_0)^\top \Delta x  + r(x_0) \|^2 \text{ subject to } \| \Delta x \| \leq \mu.
$$
It can be shown that when the constraint is active, one can find a $\lambda$ such that 
$$
\left(\nabla r(x_0) \nabla r(x_0)^\top + \lambda I\right)  \Delta x_1 = - \nabla r(x_0)^\top   r(x_0)
$$
yields a solution. Here, we do not try to find the exact $\lambda$ but keep it fixed.

```python
@printf "i=%d, f=%1.2f |df|=%1.1e \n" 0 f(x0) norm(df(x0))
x = copy(x0)
mu = 10;
lambda = 1e-4
d2f(x) = (dr(x)*dr(x)')
for i=1:10
    dx = -((d2f(x))\df(x))
    
    if norm(dx)>mu
        dx = -((d2f(x)+lambda*eye(length(x)))\df(x))
    end
#     println(norm(dx))
    
    # find out how far to go
    tt =linspace(0,1,201)
    ft = [f(x+tt[k]*dx) for k=1:length(tt)]
    id = indmin(ft)
    t0 = tt[id]
    x = x+t0*dx
    @printf "i=%d, f=%1.2f |df|=%1.1e, t=%1.1e \n" i f(x) norm(df(x)) t0
end
```

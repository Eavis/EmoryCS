## Gradient Descent for quadratic objective functions

In this example, we aim to solve
$$
    \min_{x\in\mathbb{R}^2} \frac{1}{2} x^T Q x + x^T b - c
$$
where $A \in \mathbb{R}^{2\times 2}$ is positive definite  $b \in \mathbb{R^2}$ and $c \in \mathbb{R}$. Of course, we know that the unique global minimizer is
$$
    x^* = - Q^{-1} b.
$$
Here, we want to study the performance of the gradient descent method depending on the properties of $Q$.

```python
subplot(1,2,1)
gamma = 2
Q = [gamma 0; 0 1]; b = zeros(2); c = 0.0;
x1 = linspace(-gamma,gamma,101)
x2 = linspace(-gamma,gamma,101);

f(x) = 0.5*x'*Q*x + x'*b + c
xstar = -(Q\b)
ft = [f([x1[k],x2[j]])[1] for k=1:length(x1), j=1:length(x2)]
contour(x1,x2,ft,hold=true)
plot(xstar[1],xstar[2],"or")
legend({"xstar"})
axis("equal")
title("gamma=2")

subplot(1,2,2)
gamma = 0.1
Q = [gamma 0; 0 1]; b = zeros(2); c = 0.0;

f(x) = 0.5*x'*Q*x + x'*b + c
xstar = -(Q\b)
ft = [f([x1[k],x2[j]])[1] for k=1:length(x1), j=1:length(x2)]
contour(x1,x2,ft,hold=true)
plot(xstar[1],xstar[2],"or")
legend({"xstar"})
axis("equal")
title("gamma=0.1")
```

## Computing the exact Line Search

First, let us compute the exact line search for a quadratic function $f$ at the iterate $x$. Let $d$ be a descent direction, i.e., 
$$
    \nabla f(x)^T d < 0.
$$
Then the line search function is given by
$$
    \phi: R^+ \to R, \text{ with } \phi(t) = f(x+td)
$$
This is also a quadratic, i.e., 
$$
    \phi(t) = \frac{1}{2} (x+td)^T Q (x+td) + (x+td)^T b + c =\frac{1}{2} (d^T Q d) t^2 + (d^T Q x + d^T b) t + f(x).
$$
Since $Q$ is positive definite, $\phi$ has a global minimizer that we can find by setting $\phi'$ to zero
$$
    \phi'(t) = (d^T Q d) t + d^T (Q x + b) \Rightarrow t^* = - \frac{d^T \nabla f(x)}{d^T Q d},
$$
where we have used that $\nabla f(x) = Qx + b$. Since $d$ is a descent direction, the nominator is negative and the denominator is positive since $Q$ is positive definite. So, $t^*$ is positive.

## Coding Gradient Descent

```python
function gd(f::Function,df::Function,x0,linesearch::Function;maxIter=10,tol=1e-4,out=1)
    x = copy(x0)
    his = zeros(maxIter,2)
    X   = zeros(maxIter,2)
    if out==1
        @printf "%s\t%6s\t\t%6s\n" "iter" "f(x)" "|df(x)|"
    end
    for iter=1:maxIter
        d = -df(x)
        t = linesearch(d,df(x))
        x += t*d
        his[iter,:] = [f(x) norm(df(x))];
        if out==1
            @printf "%4d\t%1.2e\t%1.2e\n" iter his[iter,1] his[iter,2]
        end
        X[iter,:] = x;
        if norm(df(x))<tol
            his = his[1:iter,:]
            X   = X[1:iter,:]
            break
        end
        
    end
    return x,his,X
end
```

```python
gamma = 0.1
Q = [gamma 0; 0 1]; b = zeros(2); c = 0.0;

x1 = linspace(-gamma,gamma,101)
x2 = linspace(-gamma,gamma,101);
f(x) = 0.5*x'*Q*x + x'*b + c
df(x) = Q*x + b
ls(d,df) = - dot(d,df)/dot(d,Q*d)

x0 = [1; gamma]
#x0 = [1;0]
xgd,his,X = gd(f,df,x0,ls,maxIter=40)

contour(x1,x2,ft,hold=true)
plot(x0[1],x0[2],"ob")
plot(X[:,1],X[:,2],"ob")
plot(xstar[1],xstar[2],"or")
axis("equal")
```

The above plot shows the first ten iterates of the Gradient Descent method (blue dots). The iterates approach the true solution (red dot), but you should also observe a zig-zag behavior. This means that the iterates flip left and right the $x_2$ axis. 

Experiment: Play with the value of $\gamma$ and observe the effect on the iterates, but also on the contour lines.

## Convergence for different gamma

From the part above, you should see that the convergence on the gradient method depends on the value of $\gamma$. Here we consider three choices
1. $\gamma=1$
2. $\gamma=0.01$
3. $\gamma=1000$

```python
gamma = 1
Q = [gamma 0; 0 1]; b = zeros(2); c = 0.0;
f(x) = 0.5*x'*Q*x + x'*b + c
df(x) = Q*x + b
ls(d,df) = - dot(d,df)/dot(d,Q*d)
x0 = [1; gamma]
xgd1,his1,X1 = gd(f,df,x0,ls,out=0)

gamma = 0.01
Q = [gamma 0; 0 1]; b = zeros(2); c = 0.0;
f(x) = 0.5*x'*Q*x + x'*b + c
df(x) = Q*x + b
ls(d,df) = - dot(d,df)/dot(d,Q*d)
x0 = [1; gamma]
xgd2,his2,X2 = gd(f,df,x0,ls,maxIter=1000; tol=1e-10,out=0);

gamma = 1000
Q = [gamma 0; 0 1]; b = zeros(2); c = 0.0;
f(x) = 0.5*x'*Q*x + x'*b + c
df(x) = Q*x + b
ls(d,df) = - dot(d,df)/dot(d,Q*d)
x0 = [1; gamma]
xgd3,his3,X3 = gd(f,df,x0,ls,maxIter=1000; tol=1e-10,out=0);
```

Good news: for $\gamma=1$ the method converges in one iteration to the exact solution!

Bad news: convergence for very small and very large $\gamma$ seems to be painfully slow. Here is the convergence:


```python
subplot(1,2,1)
semilogy( sqrt(sum(X2.^2,2)))
xlabel("iteration")
ylabel("|x - x_true|")
title("gd for \gamma=0.01")
subplot(1,2,2)
semilogy( sqrt(sum(X3.^2,2)))
xlabel("iteration")
title("gd for \gamma=1000")
```

## Estimating convergence

As we will see in the lecture, the convergence of the Gradient Descent method using exact line search for quadratic problems if given by
$$
    f(x_{k+1}) \leq \left( \frac{ \lambda_{max}(A) -  \lambda_{\min}(Q) } { \lambda_{max}(A) +  \lambda_{\min}(Q)}\right)^2 f(x_k)
$$
and so the function values decay like a geometric series. The rate of convergence depends on the relation between the largest eigenvalue of $Q$, $\lambda_{\max}(Q)$, and the smalles eigenvalue of $Q$, $\lambda_{\min}(Q)$.

Let us verify this bound for our problem.

```python
plot(1:size(his2,1)-1,his2[2:end,1]./his2[1:end-1,1],"-r",hold=true)
plot(1:size(his2,1)-1,ones(size(his2,1)-1)*((1-0.01)/(1+0.01))^2,"--k")
legend(("actual reduction","upper bound"))
xlabel("iteration");
ylabel("relative reduction")
```

This result is comforting. The actual reduction (red solid) achieves the upper bound (black dashed). In this case, we say the upper bound is tight, that is, it cannot be improved.

## Summary

This example shows a few characteristics of the Gradient Method. 
1. We have seen that the performance can be great if the condition number of $Q$ is small (i.e. the contour lines are close to a circle)
2. The convergence is really slow, when the condition number of $Q$ becomes large. 
3. The convergence crucially depends on the starting guess. Had we picked $x_0 = (1,0)^T$ the Gradient Descent method would have converged in one iteration.
